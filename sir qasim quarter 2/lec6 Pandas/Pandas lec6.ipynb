```python
def grading_system(s1,s2,s3,s4,s5):
    a=s1+s2+s3+s4+s5
    b=500
    perc=a/b*100
    return perc
```

```python
def grade_assign(s1,s2,s3,s4,s5):
    a=s1+s2+s3+s4+s5
    b=500
    perc=a/b*100
    grade=None
    if(perc>=80):
        grade='A+'
    elif(perc>=70):
        grade='A'
    elif(perc>=60):
        grade='B'
    elif(perc>=50):
        grade='C'
    elif(perc>=40):
        grade='D'
    else:
        grade='F'
    return grade
```

```python
df['perc']=df[['s1','s2','s3','s4','s5']].apply(lambda x:grading_system(x[0],x[1],x[2],x[3],x[4]),axis=1)
```

```python
df['Grade']=df[['s1','s2','s3','s4','s5']].apply(lambda x:grade_assign(x[0],x[1],x[2],x[3],x[4]),axis=1)
```
