```python
#Overall Customers
dict1={'id':[11,22,33,44],'Name':['Talha','Hamza','Ali','Zain'],'Address':[111,222,333,444]}
a=pd.DataFrame(dict1)
display(a)
```

```python
#2023 Customers
dict2={'id':[11,33,88,67,11],'order_id':[11,22,33,44,55],'order_date':[datetime.date(2023,2,1),datetime.date(2023,2,1),datetime.date(2023,1,4),datetime.date(2023,6,18),datetime.date(2023,5,1)]}
b=pd.DataFrame(dict2)
display(b)
```

```python
inner=pd.merge(b,a)
display(inner)# b will find in a
```

```python
outer=pd.merge(b,a,how='outer')
display(outer)
```

```python
left=pd.merge(b,a,how='left')
display(left)
```

```python
right=pd.merge(b,a,how='right')
display(right)
```
